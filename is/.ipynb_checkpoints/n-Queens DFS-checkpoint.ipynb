```python
class Queens:
  def __init__(self,n):
    self.n = n
  
  def isConsistent(self, board, n):
    for i in range(n):
      if (board[i] == board[n]):
        return False;   # same column
      
      if ((board[i] - board[n]) == (n - i)):
        return False   # same major diagonal
    
      if ((board[n] - board[i]) == (n - i)):
        return False   # same minor diagonal
        
    return True;
  
  def solve(self):
    board = [None]*self.n
    self.findSpot(board,0)
    
  def findSpot(self,board,k):
    if k is self.n:
      self.printBoard(board)
    else:
      for i in range(self.n):
        board[k] = i
        #print("Start: ",board)
        if self.isConsistent(board,k):
          self.findSpot(board,k+1)
        #print("\tEnd: ",board)
  
  def printBoard(self,board):
    for i in range(self.n):
      for j in range(self.n):
        if (board[i] == j):
          print("Q ", end="")
        else:
          print("* ",end="");
      print()
    print()
    
```

```python
fourQueens = Queens(4)
```

```python
fourQueens.solve()
```

```python
fiveQueens = Queens(5)
fiveQueens.solve()
```
